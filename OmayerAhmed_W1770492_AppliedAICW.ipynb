#### Tokenization (NLP)

Tokenization is a process used within NLP in order to break down raw text, symbols, phrases or any other possible elements within what has been submitted. This works within chatbots as this process breaks down customer requests/enquiries and breaks down each and every word within the phrase. Below i will provide an example of an sentance being tokenized.

```python
# Creating a function which takes in customer input 
def tokenize_chatbot(str_custinput):
    wordtokens = nltk.word_tokenize(str_custinput) 
    #this uses nltk libary (punkt) totokenize input
    
    return wordtokens
#customer input   
str_custinput = "How much are Sony XH100 wireless earbuds?"
#tokenizes customer input and prints
wordtokens = tokenize_chatbot(str_custinput)
print (wordtokens)
```

As we can see, the phrase has been tokenized into each word which then on will be stemmed.

#### Stemming

Stemming is the second process when preprocessing raw text. Once the phrase has been tokenized, stemming is a process of removing/reducing certain words that are redundant and also making associations with the rest of the words. This is used within chatbots in order to get closer to the root of the question and is able to provide an appropiate response such as if a user ask how much is of a certain product, stemming will allow the process of making associations such as "much" and be able to provide a monetary value. Below is an example of the process of stemming.

```python
def stem_tokens_example(wordtokens): #wordtoken is a variable for tokenized words
    stemmer = PorterStemmer() #from nltk lib uses to stem phrases
    
    example_stem_tokens = [stemmer.stem(wordtokens) for wordtokens in wordtokens]
    return example_stem_tokens

example_stem_tokens = stem_tokens_example(wordtokens)
print(example_stem_tokens)
```

#### Lemmatizing

Lastly, we have the lemmatizing process and is also used within NLP. when lemmatizing phrases, it allows words to be grouped together as well as have a meaning that can be understood by the program. Chatbots use this feature within NLP in order to have an actual understanding of what the phrase could mean.

```python
ExampleWords = ["How", "Much", "Are", "Soni", "xh100", "wireless", "earbud"]
lemmatizer = WordNetLemmatizer()

ExampleLemmatizer = [lemmatizer.lemmatize(ExampleWords) for ExampleWords in ExampleWords]
print(ExampleLemmatizer)
print("Here are the lemmatizer words")
```

Comapring the input and output between the three NLP proceeses, they both are unique in their own indvidual way as they are all methods which helps allow the NLP process to understand phrases and sentance structure. However, if comparing all three processes, stemming is a fast and more effective way of reducing the dimensionality of raw text however compared to lemmatization, it is a longer process as this process would be considered more accurate than stemming. As well as that, Lemmatization uses inflectional variances therefore when analysing key words within a sentance, it will take into acknowledgment of the grammatical and context use of a sentance, therefore showcasing how more accurate lemmatization is compared to stemming. As stemming uses derivational variance therefore would not take into account of context. For example if the word "Eat" was used, if lemmatized, the process would show ate or eaten which are the past and present tense. However with stemming, it would showcase words that are similar or associated such as "eating".

```python
print("Importing Training Data.....")
#Importing the JSON File into python.
#This file will be used to train the chatbot
with open ('trainingdata.json', "r") as f:
    trainingdata = json.load(f)
```

```python
#creating empty dicts to store intent
words = []
#classes for intents
classes = []
#ignore symbols within user input
ignore_symbols = ["?", "!"]
#documents
documents=[]

for intent in trainingdata["intents"]:
    for pattern in intent ["patterns"]:
        a = nltk.word_tokenize(pattern)
        words.extend(a)
        documents.append((a, intent["tag"]))
        if intent["tag"] not in classes:
            classes.append(intent["tag"])
        
print("Converting intents to each classes.....")
#This will go through json file and places words in these dictionaries

```

```python
print("We are removing redundants and stemming data")
stemmer = LancasterStemmer() # used to stem words
#Stemming words dictionary


words = [stemmer.stem(a.lower()) for a in words if a not in ignore_symbols]
words = sorted(list(set(words)))
#Stemming the words dict using stemmer and lowering tokenized words(A)


classes = sorted(list(set(classes)))
print (len(documents), "Documents")
print (len(classes), "Classes")
print (len(words), "Stemmed Words", words)
```

```python

#creating a training and output dict
training = []
output = []
output_empty = [0] * len(classes)

print("We are now creating the data model")

print("Training Set being produced....")
for doc in documents:
    bag  = [] #creates a bag dict (bag of words) 
    pattern_words = doc[0]
    #this will stem words
    pattern_words = [stemmer.stem(word.lower()) for word in pattern_words]
    for a in words:
        bag.append(1) if a in pattern_words else bag.append(0)
    #output will be 0 and 1 is a tag due to bag of words 
    output_row = list(output_empty)
    output_row[classes.index(doc[1])] = 1
    
    training.append([bag, output_row])
    
```

```python
random.shuffle(training)
training = np.array(training) # converting into numpy array 
# I have created an x,y variables in order to use for test/train model

x_train = list(training[:,0])
y_train = list(training[:,1])

print("Producing a test and train model")
print("Building a neural network")

#clears the graph stack
tf.compat.v1.reset_default_graph
```

```python
# Creating a Neural Network
chatbot_net = tflearn.input_data(shape=[None, len(x_train[0])]) #input layer for network
chatbot_net = tflearn.fully_connected(chatbot_net, 8) # dense layer with 8 neurons
chatbot_net = tflearn.fully_connected(chatbot_net, 8) # this is connected to above dense layer

chatbot_net = tflearn.fully_connected(chatbot_net, len(y_train[0]), activation="softmax") #output layer
chatbot_net = tflearn.regression(chatbot_net) #regression - nn model is stable to be trained
print("Training Chatbot Model....")
```

```python
chatbot_model = tflearn.DNN(chatbot_net, tensorboard_dir="tflearn_logs")

# This is a deep neural network modular
```

```python
print("Training the Model.......")

# I am training the chatbot model - N_Epoch is at 1000 making 1000 iterations
chatbot_model.fit(x_train, y_train, n_epoch=1000, batch_size=8, show_metric=True) #batch size 8
print("Saving the Model.......")
chatbot_model.save('chatbot_model.tflearn')
#chatbot has been saved

#self note - if error index list is too big - clear kernal output and restart
```

```python
# This function will tokenize the sentances
def sentance_cleanup(sentance):
    words_sentance = nltk.word_tokenize(sentance)
    words_sentance = [stemmer.stem(word.lower()) for word in words_sentance]
    return words_sentance

#Bag of word function - coversion of 0/1

def bagofwords(sentance, words, show_details=False):
    words_sentance = sentance_cleanup(sentance)
    bag = [0]*len(words)
    for s in words_sentance:
        for i,a in enumerate(words):
            if a  == s:
                bag[i] = 1
                if show_details:
                    print("Found %s" % a)
        return(np.array(bag))
    
Error_Limit = 0.25
print("Error_Limit = 0.25")

# function = returning values and tuples
def classify(sentance):
    results = chatbot_model.predict([bagofwords(sentance, words)])[0]
    results = [[i,r] for i,r in enumerate(results) if r>Error_Limit]
    results.sort(key=lambda x: x[1], reverse=True)
    list_return = []
    for r in results:
        list_return.append((classes[r[0]], r[1]))
        return list_return
    
    #chatbot responsees using intents
def chatbot_responses(sentance, userID="123", show_details=False):
    results = classify(sentance)
    if results:
        while results:
            for i in trainingdata["intents"]:
                if i["tag"] == results[0][0]:
                    return print(random.choice(i["responses"]))
                
            results.pop(0)

```
